```python
%matplotlib inline
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load the data matrix
A = np.loadtxt('mysterious_data.txt')
n,d = A.shape
print(f'The matrix A contains {n} points in dimension {d}')
```

Each row of $A$ corresponds to a datapoint.

```python
## data processing 
A=A-np.mean(A) # centering the data 
S=A.T@A ## computing the covarinace matrix.
## functions. 
def get_k_largest(k):
    "return k largest eigen_vectors and values"
    values, vectors=np.linalg.eigh(S)
    return values[(-1*k):],vectors[:,-1*k:]
def pca(k):
    "conduct pca bassed on the largest k values"
    k_largesst_values, k_largest_vectors=get_k_largest(k)
    return (A@k_largest_vectors) #nXk
```

looking at the elbow plot it is clear most of the variance is explained by the first two eigenvalues

```python
k_largesst_values, k_largest_vectors=get_k_largest(5) # looking at k largest eigen values
plt.plot(np.flip(k_largesst_values))  ## plotting them and flipping since the eigen values are in reverse order 
```

```python
out=pca(2) ## condcuting pca for 2 largest dimensions
y=out[:,0]*-1 ## setting y axis equal to inverted first dimesnion 
x=out[:,1] ## setting x axis equal to second dimension 
plt.scatter(x,y) ## making a scatter plot 
```
